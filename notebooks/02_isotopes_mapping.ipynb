# 02 - Isotopes & Watermarks mapping
No B2C ? → use PergamIsot + Briquet

```python
import pandas as pd
import leafmap

# 1. Charger les données
iso = pd.read_csv('../data/01_raw/b2c_isotopes.csv')
wm  = pd.read_csv('../data/01_raw/briquet_watermarks.csv')

# 2. Carte de base
m = leafmap.Map(center=[45, 5], zoom=5)
m.add_points_from_xy(
    iso,
    x="copy_place_lon",
    y="copy_place_lat",
    color_column="provenance_skin",
    layer_name="Isotopes"
)
m.to_html('../maps/isotopes_map.html')
print("✅ Carte générée : ../maps/isotopes_map.html")
```
