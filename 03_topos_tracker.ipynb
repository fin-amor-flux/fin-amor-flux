# 03 - Topos-Tracker : semantic drift animation
fin’amor → honor → virtus → valde (1150-1450)

```python
import pandas as pd
import leafmap

# 1. Base de drift (exemple chiffré)
drift = pd.DataFrame([
    {'word': 'fin’amor', 'year': 1150, 'lat': 44.8, 'lon': 0.6, 'score': 1.0},
    {'word': 'honor',   'year': 1280, 'lat': 48.9, 'lon': 2.3, 'score': 0.78},
    {'word': 'virtus',  'year': 1350, 'lat': 45.5, 'lon': 9.2, 'score': 0.54},
    {'word': 'valde',   'year': 1420, 'lat': 51.2, 'lon': 3.2, 'score': 0.32}
])

# 2. Carte animée
m = leafmap.Map(center=[47, 2], zoom=5)
m.add_points_from_xy(drift, x="lon", y="lat", color="score", 
                     color_column="score", layer_name="Semantic drift")
m.to_html('../maps/topos_heatmap.html')
print("✅ Topos-Tracker map → ../maps/topos_heatmap.html")
```
