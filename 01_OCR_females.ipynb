# 01 - OCR extraction : Female jongleurs (Rolls of Oleron)
Low-cost, no terminal needed

```python
import pandas as pd
from bs4 import BeautifulSoup

# 1. Charger le HTML brut
with open('../data/01_raw/rolls_oleron.html', encoding='utf-8') as f:
    soup = BeautifulSoup(f, 'html.parser')

# 2. Extraction ultra-simple
rows = []
for li in soup.select('li'):
    txt = li.get_text(strip=True).lower()
    if any(k in txt for k in ['cantadeira', 'jugleressa', 'mulier', 'maria']):
        rows.append({'raw_text': txt, 'gender_tag': 'F'})

df = pd.DataFrame(rows)
df.head()
```

```python
# 3. Sauvegarde
df.to_csv('../data/02_processed/females_oleron.csv', index=False)
print(f"âœ… {len(df)} occurrences saved")
```
